```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
plt.style.use('ggplot')
from skimage import io
from sklearn.cluster import KMeans
```

```python
img = io.imread('./ny.jpg')
ax = plt.axes(xticks=[],yticks=[])
ax.imshow(img)
```

```python
img_data = (img/255.0).reshape(-1,3)
img_data.shape
```

```python
from plot_utils import plot_utils
x = plot_utils(img_data, title="16mil colors")
x.colorSpace()
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
kmeans = MiniBatchKMeans(16).fit(img_data)
k_colors = kmeans.cluster_centers_[kmeans.predict(img_data)]
```

```python
y = plot_utils(img_data,colors=k_colors,title='reduced')
y.colorSpace()
```

```python
k_img = np.reshape(k_colors, (img.shape))
ax = plt.axes(xticks=[],yticks=[])
ax.imshow(k_img)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.set_title('original')
ax1.set_xticks([])
ax1.set_yticks([])
ax1.imshow(img)

ax2.set_title('compressed')
ax2.set_xticks([])
ax2.set_yticks([])
ax2.imshow(k_img)

plt.subplots_adjust(top=0.85)
plt.show()
```
